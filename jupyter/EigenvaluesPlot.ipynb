```python
import shadow_reconstruction
import numpy
from matplotlib import pyplot
```

```python
for l in [10,20,50,100,1000]:
    dat = []
    for a in range(0,min(l//2 + 1,10 + 1)):
        dat.append(shadow_reconstruction.symmetry_eigenvalue(l, a))
    pyplot.plot(dat, 'x--', label=str(l))

pyplot.plot([(2.0/3.0)**a for a in range(10 + 1)], '--', c='gray', label='$\infty$')

pyplot.ylabel('$c_\lambda$')
pyplot.xlabel('$\lambda_2$')
pyplot.grid()
pyplot.yscale('log')
pyplot.legend(title='V')
pyplot.savefig('eigenvalues.jpeg')
pyplot.savefig('eigenvalues.pdf')
pyplot.show()
```

```python
for l in [10,20,50,100,1000]:
    dat = []
    for a in range(0,min(l//2 + 1,10)):
        dat.append(shadow_reconstruction.symmetry_eigenvalue(l, a) - (2.0/3.0)**a)
    pyplot.plot(dat, 'x--', label=str(l))
    
pyplot.grid()
pyplot.legend(title='V')
pyplot.show()
```
